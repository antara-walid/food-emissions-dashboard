```python
import pandas as pd

# 1. Charger les données
df = pd.read_csv("Emissions_Totaux_F_Toutes_les_Données.csv", encoding='utf-8')

# 2. Filtrer pour ne garder QUE l'équivalent CO2 global
df = df[df['Élément'] == 'Émissions (CO2eq) (AR5)']

# 3. Définir les catégories du cycle de vie qu'on veut garder
categories_cycle = [
    "Changement d'utilisation des terres",
    "Farm gate",
    "Transformation des aliments",
    "Emballage alimentaire",
    "Transporte des alimentaires",
    "Commerce de détail alimentaire",
    "Consommation des ménages",
    "Évacuation des déchets des systèmes agroalimentaires"
]

# 4. Filtrer les produits
df = df[df['Produit'].isin(categories_cycle)]

# 5. Sélectionner les colonnes utiles (Code ISO du pays, Nom du produit, et les années 1990 à 2015)

annees = [f"Y{year}" for year in range(1990, 2016)]
colonnes_a_garder = ['Code zone (M49)', 'Zone', 'Produit'] + annees

df_light = df[colonnes_a_garder]

# 6. "Unpivot" (Melt) : Transformer les colonnes d'années en lignes pour D3.js
df_final = pd.melt(
    df_light,
    id_vars=['Code zone (M49)', 'Zone', 'Produit'],
    value_vars=annees,
    var_name='Annee',
    value_name='Emission'
)

# 7. Nettoyage final des colonnes
df_final['Annee'] = df_final['Annee'].str.replace('Y', '').astype(int) # Transformer 'Y1990' en entier 1990
df_final = df_final.rename(columns={
    'Code zone (M49)': 'id',
    'Zone': 'country',
    'Produit': 'stage'
})

# 8. Supprimer les lignes sans données (NaN)
df_final = df_final.dropna(subset=['Emission'])

# 9. Sauvegarder
df_final.to_csv("fao_cleaned.csv", index=False)

print(" Fichier 'fao_cleaned.csv' prêt pour Observable !")
print(df_final.head())
```
